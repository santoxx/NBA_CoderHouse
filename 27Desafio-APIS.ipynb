```python
from nba_api.stats.endpoints import LeagueDashPlayerStats
```

```python
columns = ["PLAYER_ID", "PLAYER_NAME", "TEAM_ID", "TEAM_ABBREVIATION", "AGE", "GP", "W", "L", "W_PCT", "MIN", "FGM", "FGA", "FG_PCT", "FG3M", "FG3A", "FG3_PCT", "FTM", "FTA", "FT_PCT", "OREB", "DREB", "REB", "AST", "TOV", "STL", "BLK", "BLKA", "PF", "PFD", "PTS", "PLUS_MINUS"]
```

```python
#Database goes from 1996-1997 to 2021-2022

# set the desired season and team
season = "2021-22"

# retrieve data from the API
player_stats = LeagueDashPlayerStats(season=season,per_mode_detailed="PerGame")
player_stats_2022 = player_stats.get_data_frames()[0]
player_stats_2022 = player_stats_2022[columns]
print(player_stats_2022)
```

```python
# set the desired season and team
season = "2009-10"

# retrieve data from the API
player_stats = LeagueDashPlayerStats(season=season,per_mode_detailed="PerGame")
player_stats_2010 = player_stats.get_data_frames()[0]
player_stats_2010 = player_stats_2010[columns]
```

```python
# create an empty list to store the data for each season
data = []

# loop through the desired seasons and retrieve the data
for season in range(1995, 2022):
    season_str = str(season) + '-' + str(season+1)[-2:]
    player_stats5 = LeagueDashPlayerStats(season=season_str, per_mode_detailed="PerGame")
    player_stats_df5 = player_stats5.get_data_frames()[0]
    player_stats_df5=player_stats_df5[columns]
    
    data.append(player_stats_df5)
    
    print(player_stats_df5) # view the data for each season
    
    
```

```python
# concatenate the data from each season into a single dataframe
all_seasons_data = pd.concat(data)
print(all_seasons_data)
```
