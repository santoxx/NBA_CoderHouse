```python
De los tres gráficos observados, y sus respectivos cálculos de correlación, podemos observar una asociación muy alta entre
la altura y el peso de los jugadores. Cuanto más alto sea el jugador, mayor será su peso. Y si apelamos al mismo razonamiento,
y nos dejamos llevar por el sentido común, aquellos jugadores que presenten mayor altura que la media también tendrán mejores
resultados a la hora de obtener rebotes. Allí se presente una correlación de mediana a moderada, mientras que sucede a la
inversa a la hora de analizar las asistencias por partido. La gráfica nos indica que alrededor de los 190cm de altura es donde
mayor cantidad de asistencias un jugador genera.
Esto puede deberse no solo a que el base (posición en el campo) es quien suele generar juego para sus compañeros. Suelen ser 
jugadores que aprovechan de su relativa pequeña estatura (en relación a la media de la liga) y su mayor agilidad para moverse
a lo largo del campo y generar opciones de anotación para él y sus compañeros. La correlación entre asistencias y altra es
claramente negativa. Al contrario de lo que sucede con los rebotes.
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df_World15=(df_World
.query("gp>15"))
```

```python
sns.heatmap(df_World15.corr(),annot=True,fmt=".1f",linewidth=.5,cmap="crest")
```

```python
df_World15 = df_World15.sort_values("season")
x = df_World15[['age','player_height','player_weight','pts','usg_pct','ast_pct']] #pts
y = df_World15['ast']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.5, shuffle = False)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
print("Test set score: {:.2f}".format(model.score(x_test, y_test)))
```

```python
y_train_pred=model.predict(x_train)
y_test_pred=model.predict(x_test)
```

```python
#CALCULO DE R2
print("R2 para el modeo: ", r2_score(y_train,y_train_pred))
```

```python
#CALCULO de MAE

#calculo mae para set de entrenamiento
mae_train=mean_absolute_error(y_train,y_train_pred)
print(mae_train)
```

```python
#calculo mae para set de prueba
mae_test= mean_absolute_error(y_test,y_test_pred)
print(mae_test)
```
